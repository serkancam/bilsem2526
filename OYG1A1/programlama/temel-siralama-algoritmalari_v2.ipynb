## 1.  Bubble Sort (KabarcÄ±k SÄ±ralamasÄ±)

### ğŸ”¹ NasÄ±l Ã‡alÄ±ÅŸÄ±r?

Listeyi baÅŸtan sona tekrar tekrar gezer. Her geÃ§iÅŸte, **yan yana duran iki elemanÄ±** karÅŸÄ±laÅŸtÄ±rÄ±r ve eÄŸer yanlÄ±ÅŸ sÄ±radalarsa (soldaki > saÄŸdaki) yerlerini deÄŸiÅŸtirir.

Her tam geÃ§iÅŸin sonunda, o anki en bÃ¼yÃ¼k eleman "kabarcÄ±k gibi" listenin en saÄŸÄ±na (doÄŸru yerine) yerleÅŸmiÅŸ olur.

### ğŸ”¹ AdÄ±m AdÄ±m Ã–rnek (`Dizi = [44, 11, 25, 12, 22, 34, 2]`)

**GeÃ§iÅŸ 1 (i=0):** (En bÃ¼yÃ¼ÄŸÃ¼ (44) en sona at)
* `[**44**, **11**], ...` -> DeÄŸiÅŸtir: `[11, 44, 25, 12, 22, 34, 2]`
* `[11, **44**, **25**], ...` -> DeÄŸiÅŸtir: `[11, 25, 44, 12, 22, 34, 2]`
* `[11, 25, **44**, **12**], ...` -> DeÄŸiÅŸtir: `[11, 25, 12, 44, 22, 34, 2]`
* `[11, 25, 12, **44**, **22**], ...` -> DeÄŸiÅŸtir: `[11, 25, 12, 22, 44, 34, 2]`
* `[11, 25, 12, 22, **44**, **34**], 2` -> DeÄŸiÅŸtir: `[11, 25, 12, 22, 34, 44, 2]`
* `[... 34, **44**, **2**]` -> DeÄŸiÅŸtir: `[11, 25, 12, 22, 34, 2, 44]`
*GeÃ§iÅŸ 1 Sonu:* `[11, 25, 12, 22, 34, 2, **44**]`

**GeÃ§iÅŸ 2 (i=1):** (Ä°kinci en bÃ¼yÃ¼ÄŸÃ¼ (34) sona at)
* `[**11**, **25**], ...` -> HayÄ±r.
* `[11, **25**, **12**], ...` -> DeÄŸiÅŸtir: `[11, 12, 25, 22, 34, 2, 44]`
* `[11, 12, **25**, **22**], ...` -> DeÄŸiÅŸtir: `[11, 12, 22, 25, 34, 2, 44]`
* `[... 22, **25**, **34**], ...` -> HayÄ±r.
* `[... 25, **34**, **2**], 44` -> DeÄŸiÅŸtir: `[11, 12, 22, 25, 2, 34, 44]`
*GeÃ§iÅŸ 2 Sonu:* `[11, 12, 22, 25, 2, **34**, **44**]`

**GeÃ§iÅŸ 3 (i=2):**
* `[**11**, **12**], ...` -> HayÄ±r.
* `[11, **12**, **22**], ...` -> HayÄ±r.
* `[11, 12, **22**, **25**], ...` -> HayÄ±r.
* `[... 22, **25**, **2**], 34, 44` -> DeÄŸiÅŸtir: `[11, 12, 22, 2, 25, 34, 44]`
*GeÃ§iÅŸ 3 Sonu:* `[11, 12, 22, 2, **25**, **34**, **44**]`

**GeÃ§iÅŸ 4 (i=3):**
* `...[11, 12, **22**, **2**], ...` -> DeÄŸiÅŸtir: `[11, 12, 2, 22, 25, 34, 44]`
*GeÃ§iÅŸ 4 Sonu:* `[11, 12, 2, **22**, **25**, **34**, **44**]`

**GeÃ§iÅŸ 5 (i=4):**
* `...[11, **12**, **2**], ...` -> DeÄŸiÅŸtir: `[11, 2, 12, 22, 25, 34, 44]`
*GeÃ§iÅŸ 5 Sonu:* `[11, 2, **12**, **22**, **25**, **34**, **44**]`

**GeÃ§iÅŸ 6 (i=5):**
* `[**11**, **2**], ...` -> DeÄŸiÅŸtir: `[2, 11, 12, 22, 25, 34, 44]`
*GeÃ§iÅŸ 6 Sonu:* `[2, 11, 12, 22, 25, 34, 44]`

*(Sonraki geÃ§iÅŸte `swapped=False` kalÄ±r ve optimizasyon dÃ¶ngÃ¼yÃ¼ kÄ±rar.)*

**SÄ±ralÄ± Dizi: `[2, 11, 12, 22, 25, 34, 44]`**

###  Python Implementasyonu 


```python
def bubble_sort_while(arr):
    n = len(arr)
    
    # DÄ±ÅŸ dÃ¶ngÃ¼ 'for i in range(n - 1)' yerine:
    i = 0
    while i < n - 1:
        swapped = False 
        
        # Ä°Ã§ dÃ¶ngÃ¼ 'for j in range(0, n - i - 1)' yerine:
        j = 0
        while j < n - i - 1:
            if arr[j] > arr[j + 1]:
                # Evrensel 'temp' deÄŸiÅŸkenli swap
                temp = arr[j]
                arr[j] = arr[j + 1]
                arr[j + 1] = temp
                swapped = True
            j += 1 # Ä°Ã§ dÃ¶ngÃ¼ sayacÄ±nÄ± artÄ±r
        
        if not swapped:
            break # Optimizasyon: DÄ±ÅŸ 'while' dÃ¶ngÃ¼sÃ¼nÃ¼ kÄ±r
            
        i += 1 # DÄ±ÅŸ dÃ¶ngÃ¼ sayacÄ±nÄ± artÄ±r
        
    return arr

# Ã–rnek KullanÄ±m:
sayilar = [44, 11, 25, 12, 22, 34, 2]
print(f"SÄ±rasÄ±z Liste: {sayilar}")
sirali_liste = bubble_sort_while(sayilar.copy()) 
print(f"SÄ±ralÄ± Liste: {sirali_liste}")

sayilar2 = [64, 34, 25, 12, 22, 11, 90]
print(f"\nDiÄŸer SÄ±rasÄ±z Liste: {sayilar2}")
print(f"DiÄŸer SÄ±ralÄ± Liste: {bubble_sort_while(sayilar2.copy())}")
```

## 2.  Selection Sort (SeÃ§meli SÄ±ralama)

### ğŸ”¹ NasÄ±l Ã‡alÄ±ÅŸÄ±r?

Listeyi "sÄ±ralÄ±" ve "sÄ±rasÄ±z" olmak Ã¼zere iki sanal bÃ¶lÃ¼me ayÄ±rÄ±r.

1.  Listenin **sÄ±rasÄ±z** bÃ¶lÃ¼mÃ¼ndeki **en kÃ¼Ã§Ã¼k** elemanÄ± bulur.
2.  Bu en kÃ¼Ã§Ã¼k elemanÄ±, sÄ±rasÄ±z bÃ¶lÃ¼mÃ¼n **en baÅŸÄ±ndaki** eleman ile yer deÄŸiÅŸtirir.
3.  SÄ±ralÄ± bÃ¶lÃ¼mÃ¼n sÄ±nÄ±rÄ±nÄ± bir eleman saÄŸa kaydÄ±rÄ±r.

### ğŸ”¹ AdÄ±m AdÄ±m Ã–rnek (`Dizi = [44, 11, 25, 12, 22, 34, 2]`)

**GeÃ§iÅŸ 1 (i=0):** (SÄ±ralÄ±: `[]` | SÄ±rasÄ±z: `[44, 11, 25, 12, 22, 34, 2]`)
* En kÃ¼Ã§Ã¼ÄŸÃ¼ ara (j=1..6): `min_index = 0` (44). `j=1` (11) -> `min_index=1`. `j=6` (2) -> `min_index=6`.
* En kÃ¼Ã§Ã¼k '2' (indeks 6).
* `arr[0]` (44) ile `arr[6]` (2) yer deÄŸiÅŸtir -> `[**2**, 11, 25, 12, 22, 34, **44**]`
*GeÃ§iÅŸ 1 Sonu.* (SÄ±ralÄ±: `[2]` | SÄ±rasÄ±z: `[11, 25, 12, 22, 34, 44]`)

**GeÃ§iÅŸ 2 (i=1):** (SÄ±ralÄ±: `[2]` | SÄ±rasÄ±z: `[11, 25, 12, 22, 34, 44]`)
* En kÃ¼Ã§Ã¼ÄŸÃ¼ ara (j=2..6): `min_index = 1` (11). Kalanlar daha bÃ¼yÃ¼k.
* En kÃ¼Ã§Ã¼k '11' (indeks 1).
* `arr[1]` (11) ile `arr[1]` (11) yer deÄŸiÅŸtir. (DeÄŸiÅŸiklik yok).
*GeÃ§iÅŸ 2 Sonu.* (SÄ±ralÄ±: `[2, 11]` | SÄ±rasÄ±z: `[25, 12, 22, 34, 44]`)

**GeÃ§iÅŸ 3 (i=2):** (SÄ±ralÄ±: `[2, 11]` | SÄ±rasÄ±z: `[25, 12, 22, 34, 44]`)
* En kÃ¼Ã§Ã¼ÄŸÃ¼ ara (j=3..6): `min_index = 2` (25). `j=3` (12) -> `min_index=3`.
* En kÃ¼Ã§Ã¼k '12' (indeks 3).
* `arr[2]` (25) ile `arr[3]` (12) yer deÄŸiÅŸtir -> `[2, 11, **12**, **25**, 22, 34, 44]`
*GeÃ§iÅŸ 3 Sonu.* (SÄ±ralÄ±: `[2, 11, 12]` | SÄ±rasÄ±z: `[25, 22, 34, 44]`)

**GeÃ§iÅŸ 4 (i=3):** (SÄ±ralÄ±: `[2, 11, 12]` | SÄ±rasÄ±z: `[25, 22, 34, 44]`)
* En kÃ¼Ã§Ã¼ÄŸÃ¼ ara (j=4..6): `min_index = 3` (25). `j=4` (22) -> `min_index=4`.
* En kÃ¼Ã§Ã¼k '22' (indeks 4).
* `arr[3]` (25) ile `arr[4]` (22) yer deÄŸiÅŸtir -> `[2, 11, 12, **22**, **25**, 34, 44]`
*GeÃ§iÅŸ 4 Sonu.* (SÄ±ralÄ±: `[2, 11, 12, 22]` | SÄ±rasÄ±z: `[25, 34, 44]`)

**GeÃ§iÅŸ 5 (i=4):** (SÄ±ralÄ±: `[...22]` | SÄ±rasÄ±z: `[25, 34, 44]`)
* En kÃ¼Ã§Ã¼ÄŸÃ¼ ara (j=5..6): `min_index = 4` (25). Kalanlar daha bÃ¼yÃ¼k.
* En kÃ¼Ã§Ã¼k '25' (indeks 4). DeÄŸiÅŸiklik yok.
*GeÃ§iÅŸ 5 Sonu.* (SÄ±ralÄ±: `[...25]` | SÄ±rasÄ±z: `[34, 44]`)

**GeÃ§iÅŸ 6 (i=5):** (SÄ±ralÄ±: `[...25]` | SÄ±rasÄ±z: `[34, 44]`)
* En kÃ¼Ã§Ã¼ÄŸÃ¼ ara (j=6..6): `min_index = 5` (34). `j=6` (44) daha bÃ¼yÃ¼k.
* En kÃ¼Ã§Ã¼k '34' (indeks 5). DeÄŸiÅŸiklik yok.
*GeÃ§iÅŸ 6 Sonu.* (SÄ±ralÄ±: `[...34]` | SÄ±rasÄ±z: `[44]`)

*(Son geÃ§iÅŸ (i=6) Ã§alÄ±ÅŸÄ±r, `j` dÃ¶ngÃ¼sÃ¼ Ã§alÄ±ÅŸmaz, `arr[6]` (44) kendisiyle deÄŸiÅŸir, biter.)*

**SÄ±ralÄ± Dizi: `[2, 11, 12, 22, 25, 34, 44]`**

###  Python Implementasyonu


```python
def selection_sort_while(arr):
    n = len(arr)
    
    # DÄ±ÅŸ dÃ¶ngÃ¼ 'for i in range(n)' yerine:
    i = 0
    while i < n:
        min_index = i
        
        # Ä°Ã§ dÃ¶ngÃ¼ 'for j in range(i + 1, n)' yerine:
        j = i + 1
        while j < n:
            if arr[j] < arr[min_index]:
                min_index = j
            j += 1 # Ä°Ã§ dÃ¶ngÃ¼ sayacÄ±nÄ± artÄ±r
        
        # Swap iÅŸlemi
        temp = arr[i]
        arr[i] = arr[min_index]
        arr[min_index] = temp
        
        i += 1 # DÄ±ÅŸ dÃ¶ngÃ¼ sayacÄ±nÄ± artÄ±r
    
    return arr

# Ã–rnek KullanÄ±m:
sayilar = [44, 11, 25, 12, 22, 34, 2]
print(f"SÄ±rasÄ±z Liste: {sayilar}")
sirali_liste = selection_sort_while(sayilar.copy()) 
print(f"SÄ±ralÄ± Liste: {sirali_liste}")

sayilar2 = [64, 34, 25, 12, 22, 11, 90]
print(f"\nDiÄŸer SÄ±rasÄ±z Liste: {sayilar2}")
print(f"DiÄŸer SÄ±ralÄ± Liste: {selection_sort_while(sayilar2.copy())}")
```

## 3.  Insertion Sort (Araya Ekleme SÄ±ralamasÄ±)

### ğŸ”¹ NasÄ±l Ã‡alÄ±ÅŸÄ±r?

MantÄ±ÄŸÄ± elinizdeki oyun kartlarÄ±nÄ± sÄ±ralamaya benzer. Liste "sÄ±ralÄ±" ve "sÄ±rasÄ±z" olarak ikiye bÃ¶lÃ¼nÃ¼r (BaÅŸlangÄ±Ã§ta ilk eleman 'sÄ±ralÄ±' kabul edilir).

1.  SÄ±rasÄ±z bÃ¶lÃ¼mÃ¼n ilk elemanÄ±nÄ± (buna `key` diyelim) elinize alÄ±n.
2.  Bu `key` elemanÄ±nÄ±, solundaki **sÄ±ralÄ±** bÃ¶lÃ¼mdeki elemanlarla (saÄŸdan sola) tek tek karÅŸÄ±laÅŸtÄ±rÄ±n.
3.  `key`'den bÃ¼yÃ¼k olan tÃ¼m elemanlarÄ± bir pozisyon saÄŸa **kaydÄ±rÄ±n**.
4.  `key` elemanÄ± iÃ§in doÄŸru pozisyon bulunduÄŸunda, oraya "araya ekleyin".

### ğŸ”¹ AdÄ±m AdÄ±m Ã–rnek (`Dizi = [44, 11, 25, 12, 22, 34, 2]`)

**BaÅŸlangÄ±Ã§:** (SÄ±ralÄ±: `[44]` | SÄ±rasÄ±z: `[11, 25, 12, 22, 34, 2]`)

**GeÃ§iÅŸ 1 (i=1):** (key = 11)
* `key (11) < 44` mÃ¼? Evet. 44'Ã¼ saÄŸa kaydÄ±r: `[44, 44, 25, ...]`
* `j` -1 oldu. 11'i `j+1` (0)'a ekle: `[**11**, 44, 25, 12, 22, 34, 2]`
*Sonu:* (SÄ±ralÄ±: `[11, 44]` | SÄ±rasÄ±z: `[25, 12, 22, 34, 2]`)

**GeÃ§iÅŸ 2 (i=2):** (key = 25)
* `key (25) < 44` mÃ¼? Evet. 44'Ã¼ saÄŸa kaydÄ±r: `[11, 44, 44, 12, ...]`
* `key (25) < 11` mÃ¼? HayÄ±r. Dur.
* 25'i `j+1` (1)'e ekle: `[11, **25**, 44, 12, 22, 34, 2]`
*Sonu:* (SÄ±ralÄ±: `[11, 25, 44]` | SÄ±rasÄ±z: `[12, 22, 34, 2]`)

**GeÃ§iÅŸ 3 (i=3):** (key = 12)
* `key (12) < 44` mÃ¼? Evet. KaydÄ±r: `[11, 25, 44, 44, 22, ...]`
* `key (12) < 25` mÃ¼? Evet. KaydÄ±r: `[11, 25, 25, 44, 22, ...]`
* `key (12) < 11` mÃ¼? HayÄ±r. Dur.
* 12'yi `j+1` (1)'e ekle: `[11, **12**, 25, 44, 22, 34, 2]`
*Sonu:* (SÄ±ralÄ±: `[11, 12, 25, 44]` | SÄ±rasÄ±z: `[22, 34, 2]`)

**GeÃ§iÅŸ 4 (i=4):** (key = 22)
* `key (22) < 44` mÃ¼? Evet. KaydÄ±r: `[... 25, 44, 44, 34, 2]`
* `key (22) < 25` mÃ¼? Evet. KaydÄ±r: `[... 12, 25, 25, 44, 34, 2]`
* `key (22) < 12` mÃ¼? HayÄ±r. Dur.
* 22'yi `j+1` (2)'ye ekle: `[11, 12, **22**, 25, 44, 34, 2]`
*Sonu:* (SÄ±ralÄ±: `[11, 12, 22, 25, 44]` | SÄ±rasÄ±z: `[34, 2]`)

**GeÃ§iÅŸ 5 (i=5):** (key = 34)
* `key (34) < 44` mÃ¼? Evet. KaydÄ±r: `[... 25, 44, 44, 2]`
* `key (34) < 25` mÃ¼? HayÄ±r. Dur.
* 34'Ã¼ `j+1` (4)'e ekle: `[11, 12, 22, 25, **34**, 44, 2]`
*Sonu:* (SÄ±ralÄ±: `[11, 12, 22, 25, 34, 44]` | SÄ±rasÄ±z: `[2]`)

**GeÃ§iÅŸ 6 (i=6):** (key = 2)
* `key (2) < 44` mÃ¼? Evet. KaydÄ±r. `... 44, 44]`
* `key (2) < 34` mÃ¼? Evet. KaydÄ±r. `... 34, 44, 44]`
* `key (2) < 25` mÃ¼? Evet. KaydÄ±r. `... 25, 34, 44, 44]`
* `key (2) < 22` mÃ¼? Evet. KaydÄ±r. `... 22, 25, 34, 44, 44]`
* `key (2) < 12` mÃ¼? Evet. KaydÄ±r. `... 12, 22, 25, 34, 44, 44]`
* `key (2) < 11` mÃ¼? Evet. KaydÄ±r: `[11, 11, 12, 22, 25, 34, 44]`
* `j` -1 oldu. Dur.
* 2'yi `j+1` (0)'a ekle: `[**2**, 11, 12, 22, 25, 34, 44]`
*Sonu:* (SÄ±ralÄ±: `[2, 11, 12, 22, 25, 34, 44]` | SÄ±rasÄ±z: `[]`)

**SÄ±ralÄ± Dizi: `[2, 11, 12, 22, 25, 34, 44]`**

###  Python Implementasyonu 
*(Not: Bu algoritmada 'swap' yerine 'kaydÄ±rma' (shifting) olduÄŸu iÃ§in temp deÄŸiÅŸkenine gerek yoktur, 'key' deÄŸiÅŸkeni bu iÅŸi gÃ¶rÃ¼r.)*

```python
def insertion_sort_while(arr):
    n = len(arr)
    
    # DÄ±ÅŸ dÃ¶ngÃ¼ 'for i in range(1, n)' yerine:
    i = 1
    while i < n:
        key = arr[i] # Araya eklenecek olan eleman (temp gibi davranÄ±r)
        
        # Ä°Ã§ dÃ¶ngÃ¼ (zaten 'while' idi):
        # 'key'i, solundaki sÄ±ralÄ± bÃ¶lÃ¼mle karÅŸÄ±laÅŸtÄ±rÄ±r
        j = i - 1
        
        while j >= 0 and key < arr[j]:
            # ElemanÄ± bir saÄŸa kaydÄ±r (yer aÃ§mak iÃ§in)
            arr[j + 1] = arr[j]
            j -= 1 # Bir Ã¶nceki elemana bakmak iÃ§in sola git
        
        # 'key'i doÄŸru yerine (j'nin bir saÄŸÄ±na) yerleÅŸtir.
        arr[j + 1] = key
        
        i += 1 # DÄ±ÅŸ dÃ¶ngÃ¼ sayacÄ±nÄ± artÄ±r
    
    return arr

# Ã–rnek KullanÄ±m:
sayilar = [44, 11, 25, 12, 22, 34, 2]
print(f"SÄ±rasÄ±z Liste: {sayilar}")
sirali_liste = insertion_sort_while(sayilar.copy()) 
print(f"SÄ±ralÄ± Liste: {sirali_liste}")

sayilar2 = [64, 34, 25, 12, 22, 11, 90]
print(f"\nDiÄŸer SÄ±rasÄ±z Liste: {sayilar2}")
print(f"DiÄŸer SÄ±ralÄ± Liste: {insertion_sort_while(sayilar2.copy())}")
```
