```python
# 1-15 arasındaki sayıların toplamını bulunuz.
# bilgisayar yazılımlarında eğer algoritmik çözümler varsa, bunların 3 türü vardır(klasik algoritmalar için)
# 1- tek adımda formül ile
# 2- döngüler ile
# 3- öz yineleme ile(bu kısım bir çok geniş bir yönretmler topluluğudur diyebiliriz.)
```

```python
#1. yol
n=15
toplam=n*(n+1)/2
print(f"1-15 arası sayıları toplamı:{toplam}")
```

```python
#2. yol
i=1
son=15
toplam=0
while i<=son:
    toplam=toplam+i
    i=i+1
print(toplam)
```

```python
# while döngüsü ile kullanıcıdan alınan iki sayı arasındaki çift sayıların toplamını bulunuz.
ilk=int(input("küçük sayıyı giriniz:"))
son=int(input("büyük sayıyı giriniz:"))
#döngü değişkeni t
t=ilk
toplam=0
while t<=son:
    # print(t)
    if t%2==0:
        # print(t)
        toplam=toplam+t
    t=t+1

print(f"toplam:{toplam}")
```

```python
# iç içe while döngüleri
#aşağıdaki gibi çarpım tablosunun birler ikiler ve üçler için yazdıralım
"""
1x1=1
1x2=2
1
...
1x10=10
...
3x10=30
"""
s1=1
s2=1
while s1<=3:
    s2=1
    while s2<=10:
        print(f"{s1}x{s2}={s1*s2}")
        s2+=1
    s1+=1

```
