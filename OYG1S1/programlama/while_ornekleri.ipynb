```python
toplam=0
n=20
dd=1
while dd<=n: # dd değişkeninin 1 - 2 - 3 - 4 - 5 .... n  durumlarında True n+1 olunca False
    toplam=toplam+dd
    dd+=1
print(toplam)
```

```python
# while döngüsü ile kullanıcıdan alınan iki sayı arasındaki çift sayıların toplamını bulunuz.
ilk=int(input("ilk sayıyı ghiriniz:"))
son=int(input("ikinci ghiriniz:"))
t=ilk
toplam=0
while t<=son:
    # print(t)
    if t&1==0:#döngünün bu adımında t sayısı çiftmiş
        toplam=toplam+t
        
    t=t+1
print(f"{ilk}-{son} arası çift sayılar toplamı:{toplam}")
```
